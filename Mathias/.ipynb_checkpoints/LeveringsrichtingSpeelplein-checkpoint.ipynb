```python
import pandas as pd
import numpy as np
import matplotlib
data = pd.read_csv('../datasets/sicknieuwmapje/SuperLiander.csv', sep='\t')
```

```python
lev_list = []
for index, row in data.iterrows():
    lev = float(row['%Leveringsrichting'].replace(',','.'))
    lev_list.append(lev)
    
print(len(data))
print(len(lev_list))
```

```python
del data['%Leveringsrichting']
data['Lev'] = lev_list
```

```python
wijken = ['CENTRUM', 'NOORD', 'WESTPOORT', 'WEST', 'NIEUW-WEST', 'ZUID', 'OOST', 'ZUIDOOST', 'DRIEMOND']
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017]

for wijk in wijken:
    for year in years:
        wijk_data = data[data['WIJKEN'] == wijk]
        jaar_data = wijk_data[wijk_data['jaar'] == year]
        count = jaar_data[jaar_data['Lev'] < 100]['Lev'].count()
        print('Wijk: {}, jaar: {}, aantal teruglevering: {}'.format(wijk, year, count))
        print('-'*30)
```
