Hierin ga ik proberen een bar chart te maken m het gebruik per stadsdeel over de jaren 2011 tot 2016 te plotten

```python
from bokeh.plotting import figure
from bokeh.io import output_file, show, save
import pandas as pd
import numpy as np
import math
```

```python
from bokeh.io import show, output_file
from bokeh.plotting import figure

output_file("stadsdeelgebruik-bars.html")

fruits = ['2011', '2012', '2013', '2014', '2015', '2016', '2017']

p = figure(x_range=fruits, plot_height=3000, title="Gebruik per stadsdeel per jaar")
p.vbar(x=fruits, top=[5, 3, 4, 2, 4, 6, 10], width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

save(p)
```

```python
# vervolgens delen we de data over een stadsdeel op in de bijbehorende jaren
centrum_2011 = centrum[centrum['jaar'] == 2011]  
centrum_2012 = centrum[centrum['jaar'] == 2012]  
centrum_2013 = centrum[centrum['jaar'] == 2013]  
centrum_2014 = centrum[centrum['jaar'] == 2014]  
centrum_2015 = centrum[centrum['jaar'] == 2015]  
centrum_2016 = centrum[centrum['jaar'] == 2016]  
centrum_2017 = centrum[centrum['jaar'] == 2017]  

```

```python
gem_stroom_per_jaar = {'Centrum_2011' : centrum_2011['SJV'].mean(), 'Centrum_2012' : centrum_2012['SJV'].mean(),
                       'Centrum_2013' : centrum_2013['SJV'].mean(),'Centrum_2014' : centrum_2014['SJV'].mean(),
                      'Centrum_2015' : centrum_2015['SJV'].mean(),'Centrum_2016' : centrum_2016['SJV'].mean(),
                      'Centrum_2017' : centrum_2017['SJV'].mean(),} 

print(gem_stroom_per_jaar)
```
