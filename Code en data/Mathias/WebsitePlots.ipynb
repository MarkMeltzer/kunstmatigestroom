```python
# data voor SJV plot per wijk per data(westpoort)
wijken = ['CENTRUM', 'NOORD', 'WESTPOORT', 'WEST', 'NIEUW-WEST', 'ZUID', 'OOST']
jaren = ['2011', '2012', '2013', '2014', '2015', '2016', '2017']


data3= pd.read_csv('../datasets/sicknieuwmapje/SuperLiander.csv', sep='\t', low_memory=False)
elk = data3[data3['PRODUCTSOORT'] == 'ELK']

list_2011 = []
for wijk in wijken:
    list_2011.append(elk[(elk['WIJKEN'] == wijk) & (elk['jaar'] == 2011)]['SJV'].mean())
list_2012 = []
for wijk in wijken:
    list_2012.append(elk[(elk['WIJKEN'] == wijk) & (elk['jaar'] == 2012)]['SJV'].mean())
list_2013 = []
for wijk in wijken:
    list_2013.append(elk[(elk['WIJKEN'] == wijk) & (elk['jaar'] == 2013)]['SJV'].mean())
list_2014 = []
for wijk in wijken:
    list_2014.append(elk[(elk['WIJKEN'] == wijk) & (elk['jaar'] == 2014)]['SJV'].mean())
list_2015 = []
for wijk in wijken:
    list_2015.append(elk[(elk['WIJKEN'] == wijk) & (elk['jaar'] == 2015)]['SJV'].mean())
list_2016 = []
for wijk in wijken:
    list_2016.append(elk[(elk['WIJKEN'] == wijk) & (elk['jaar'] == 2016)]['SJV'].mean())
list_2017 = []
for wijk in wijken:
    list_2017.append(elk[(elk['WIJKEN'] == wijk) & (elk['jaar'] == 2017)]['SJV'].mean())
```

```python
output_file("htmlplots\westpoort_plot.html")

data = {'wijken' : wijken,
        '2011'   : list_2011,
        '2012'   : list_2012,
        '2013'   : list_2013,
        '2014'   : list_2014,
        '2015'   : list_2015,
        '2016'   : list_2016,
        '2017'   : list_2017}

x = [ (wijk, jaar) for wijk in wijken for jaar in jaren ]
sjv = sum(zip(data['2011'], data['2012'], data['2013'], data['2014'], data['2015'], data['2016'], data['2017']), ())

source = ColumnDataSource(data=dict(x=x, sjv=sjv))

p = figure(plot_width=800, 
           plot_height=500,
           x_range=FactorRange(*x),
           tools="",
           toolbar_location="below",
           title="Gem. stroomverbruik per postcodegebied per stadsdeel per jaar")

p.sizing_mode = 'scale_width'

p.title.text_color='white'

p.background_fill_alpha = 0.5
p.border_fill_alpha = 0.35

b = p.vbar(x='x', top='sjv', width=0.8, color='indigo', source=source)

p.add_tools(HoverTool(renderers=[b], tooltips=[('Gem. stroomverbruik',"@sjv{int}")]))

p.yaxis.axis_label = "Gem. jaarstroomverbruik"

p.legend.location = "bottom_left"
p.legend.background_fill_alpha = 0.5

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 'vertical'
p.xgrid.grid_line_color = None

# show the results
save(p)
```
