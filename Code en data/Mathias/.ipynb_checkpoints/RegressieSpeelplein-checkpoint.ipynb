```python
import pandas as pd
import numpy as np
import matplotlib
data = pd.read_csv('../datasets/sicknieuwmapje/SuperLiander.csv', sep='\t')
```

```python
elk = data[data['PRODUCTSOORT'] == 'ELK']
elk11 = elk[elk['jaar'] == 2011]
elk12 = elk[elk['jaar'] == 2012]
elk13 = elk[elk['jaar'] == 2013]
elk14 = elk[elk['jaar'] == 2014]
elk15 = elk[elk['jaar'] == 2015]
elk16 = elk[elk['jaar'] == 2016]
elk17 = elk[elk['jaar'] == 2017]
```

```python
slim_list11 = []
for index, row in elk11.iterrows():
    aantal_aansl = row['Aantal Aansluitingen']
    stat_float = float(row['%Fysieke status'].replace(',','.'))/100
    slim_float = float(row['%Slimme Meter'].replace(',','.'))/100
    aantal_slim = aantal_aansl * stat_float * slim_float
    slim_list11.append(aantal_slim)
sjv_list11 = list(elk11['SJV'])

slim_list12 = []
for index, row in elk12.iterrows():
    aantal_aansl = row['Aantal Aansluitingen']
    stat_float = float(row['%Fysieke status'].replace(',','.'))/100
    slim_float = float(row['%Slimme Meter'].replace(',','.'))/100
    aantal_slim = aantal_aansl * stat_float * slim_float
    slim_list12.append(aantal_slim)
sjv_list12 = list(elk12['SJV'])

slim_list13 = []
for index, row in elk13.iterrows():
    aantal_aansl = row['Aantal Aansluitingen']
    stat_float = float(row['%Fysieke status'].replace(',','.'))/100
    slim_float = float(row['%Slimme Meter'].replace(',','.'))/100
    aantal_slim = aantal_aansl * stat_float * slim_float
    slim_list13.append(aantal_slim)
sjv_list13 = list(elk13['SJV'])

slim_list14 = []
for index, row in elk14.iterrows():
    aantal_aansl = row['Aantal Aansluitingen']
    stat_float = float(row['%Fysieke status'].replace(',','.'))/100
    slim_float = float(row['%Slimme Meter'].replace(',','.'))/100
    aantal_slim = aantal_aansl * stat_float * slim_float
    slim_list14.append(aantal_slim)
sjv_list14 = list(elk14['SJV'])

slim_list15 = []
for index, row in elk15.iterrows():
    aantal_aansl = row['Aantal Aansluitingen']
    stat_float = float(row['%Fysieke status'].replace(',','.'))/100
    slim_float = float(row['%Slimme Meter'].replace(',','.'))/100
    aantal_slim = aantal_aansl * stat_float * slim_float
    slim_list15.append(aantal_slim)
sjv_list15 = list(elk15['SJV'])

slim_list16 = []
for index, row in elk16.iterrows():
    aantal_aansl = row['Aantal Aansluitingen']
    stat_float = float(row['%Fysieke status'].replace(',','.'))/100
    slim_float = float(row['%Slimme Meter'].replace(',','.'))/100
    aantal_slim = aantal_aansl * stat_float * slim_float
    slim_list16.append(aantal_slim)
sjv_list16 = list(elk16['SJV'])

slim_list17 = []
for index, row in elk17.iterrows():
    aantal_aansl = row['Aantal Aansluitingen']
    stat_float = float(row['%Fysieke status'].replace(',','.'))/100
    slim_float = float(row['%Slimme Meter'].replace(',','.'))/100
    aantal_slim = aantal_aansl * stat_float * slim_float
    slim_list17.append(aantal_slim)
sjv_list17 = list(elk17['SJV'])
```

```python
import random
list11 = []
list12 = []
list13 = []
list14 = []
list15 = []
list16 = []
list17 = []

for i in range(len(elk11)):
    list11.append(random.uniform(2010.3,2011))

for i in range(len(elk12)):
    list12.append(random.uniform(2011.3,2012))
    
for i in range(len(elk13)):
    list13.append(random.uniform(2012.3,2013))
    
for i in range(len(elk14)):
    list14.append(random.uniform(2013.3,2014))
    
for i in range(len(elk15)):
    list15.append(random.uniform(2014.3,2015))
    
for i in range(len(elk16)):
    list16.append(random.uniform(2015.3,2016))
    
for i in range(len(elk17)):
    list17.append(random.uniform(2016.3,2017))
```

```python
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.sampledata.commits import data
from bokeh.transform import jitter
from bokeh.models import CrosshairTool
from bokeh.models import LinearAxis, Range1d

# output to static HTML file
output_file("line.html")

source = ColumnDataSource(data=dict(
    x=slim_list11,
    x2=slim_list12,
    x3=slim_list13,
    x4=slim_list14,
    x5=slim_list15,
    x6=slim_list16,
    x7=slim_list17,
    y=list11,
    y2=list12,
    y3=list13,
    y4=list14,
    y5=list15,
    y6=list16,
    y7=list17    
))

p = figure(plot_width=600, plot_height=350, y_range=(2009.5,2018), title=':)')

# add a circle renderer with a size, color, and alpha
p.circle('x', 'y', source=source, alpha=0.3, color='mediumvioletred')
p.circle('x2', 'y2', source=source, alpha=0.3, color='mediumvioletred')
p.circle('x3', 'y3', source=source, alpha=0.3, color='mediumvioletred')
p.circle('x4', 'y4', source=source, alpha=0.3, color='mediumvioletred')
p.circle('x5', 'y5', source=source, alpha=0.3, color='mediumvioletred')
p.circle('x6', 'y6', source=source, alpha=0.3, color='mediumvioletred')
p.circle('x7', 'y7', source=source, alpha=0.3, color='mediumvioletred')

p.add_tools(CrosshairTool())

p.xaxis.axis_label = "Aantal slimme meters"
p.yaxis.minor_tick_line_color = None
p.yaxis.ticker = [2011, 2012, 2013, 2014, 2015, 2016, 2017]

# show the results
show(p)
```
