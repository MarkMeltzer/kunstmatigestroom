```python
import pandas as pd
import numpy as np
import matplotlib
data = pd.read_csv('..\datasets\sicknieuwmapje\OriginalTotaal.csv', sep='\t', low_memory=False)
```

```python
data.boxplot(column=['SJV'], figsize=[5,10])
```

```python
# Filter all data using box and whiskerplot
Q1 = data['SJV'].quantile(0.25)
Q3 = data['SJV'].quantile(0.75)
IQR = Q3 - Q1

filtered_data = data.query('(@Q1 - 1.5 * @IQR) <= SJV <= (@Q3 + 1.5 * @IQR)')
```

```python
filtered_data['SJV'].hist(figsize=[20,10], bins=750)
```

```python
filtered_data_2 = filtered_data.query('450 <= SJV')
filtered_data_2['SJV'].hist(figsize=[20,10], bins=750)
```
